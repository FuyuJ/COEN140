```python
"""
Lab 3 - COEN 140
Marianne Fuyu Yamazaki Dorr
10/3/2022

This program will use Linear Discriminant Analysis (LDA) and Quadratic Discriminant Analysis (QDA) to 
analyze some data. 
"""

import numpy as np
'''import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import colors
from scipy import linalg'''

#turns data from a file into a list
def data_to_list(filename):
    dataList = []
    file = open(filename, 'r')
    
    for line in file:
        stripped_line = line.strip()
        line_list = stripped_line.split(",")
        for x in range(4):
            line_list[x] = np.float64(line_list[x])
            
        dataList.append(line_list)
    
    file.close()
    return dataList    

#provided function for testing if the data is in the proper initial format
def test_dataset(data):
    if len(data) != 150:
        print("Length is wrong")
        return False
    
    for row in data:
        if len(row) != 5:
            print("list length is wrong")
            return False
        
        for column in row[:-1]:
            if type(column) != np.float64:
                print(column)
                print("Not correct type")
                return False
            
        if type(row[-1]) != str:
            print("wrong str")
            return False
    
    return True

#split data into training set and testing set (80%/20%)
def split_data(data):
    training_data = []
    testing_data = []
    C1, C2, C3 = ([] for i in range(3))
    index = 0
    #split data into the three categories
    C1name = data[index][4]
    while (data[index][4] == C1name):
        C1.append(data[index])
        index = index + 1
    C2name = data[index][4]
    while (data[index][4] == C2name):
        C2.append(data[index])
        index = index + 1
    C3name = data[index][4]
    while (index < len(data)):
        C3.append(data[index])
        index = index + 1
    #add 80% of each category to training_data list and remaining 20% to testing_data list
    t_index1 = 0.8 * len(C1)
    t_index2 = 0.8 * len(C2)
    t_index3 = 0.8 * len(C3)
    index_category_list = [[t_index1, C1], [t_index2, C2], [t_index3, C3]]
    for c in index_category_list:
        i = 0
        while (i < c[0]):
            training_data.append(c[1][i])
            i += 1
        while (i < len(c[1])):
            testing_data.append(c[1][i])
            i += 1
    
    return training_data, testing_data
 
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis 
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis 

def LDA(training_data, test_data):
    #split data into X and y where X is the measurements of each iris and y is the name of the iris
    X_training = []
    y_training = []
    for d in training_data:
        X_training.append(d[:-1])
        y_training.append(d[-1])
    X_testing = []
    y_testing = []
    for t in test_data:
        X_testing.append(t[:-1])
        y_testing.append(t[-1])
    
    lda = LinearDiscriminantAnalysis(solver="svd", store_covariance=True)
    lda = lda.fit(X_training, y_training)
    y_pred_training = lda.predict(X_training)
    
    y_pred_test = lda.predict(X_testing)
    
    #Accuracy
    score = 0
    for i in range(len(y_training)):
        if (y_training[i] != y_pred_training[i]):
            print("Error at index: " + str(i))
            print("Actual Data: " + y_training[i])
            print("Predicted Data: " + y_pred_training[i])
            print()
        else:
            score += 1
    print("Error rate of LDA on training data set: " + str((1-(score/len(y_training)))*100) + "%") 
    print()
    
    #Accuracy
    score = 0
    for i in range(len(y_testing)):
        if (y_testing[i] != y_pred_test[i]):
            print("Error at index: " + str(i))
            print("Actual Data: " + y_testing[i])
            print("Predicted Data: " + y_pred_test[i])
            print()
        else:
            score += 1
    print("Error rate of LDA on testing data set: " + str((1-(score/len(y_testing)))*100) + "%") 
    print()
    #return y_pred_test

def QDA(training_data, testing_data):
    #split data into X and y where X is the measurements of each iris and y is the name of the iris
    X_training = []
    y_training = []
    for d in training_data:
        X_training.append(d[:-1])
        y_training.append(d[-1])
    X_testing = []
    y_testing = []
    for t in testing_data:
        X_testing.append(t[:-1])
        y_testing.append(t[-1])

    qda = QuadraticDiscriminantAnalysis(store_covariance=True)
    qda = qda.fit(X_training, y_training)
    y_pred_training = qda.predict(X_training)
    y_pred_test = qda.predict(X_testing)
    
    #Accuracy
    score = 0
    for i in range(len(y_training)):
        if (y_training[i] != y_pred_training[i]):
            print("Error at index: " + str(i))
            print("Actual Data: " + y_training[i])
            print("Predicted Data: " + y_pred_training[i])
            print()
        else:
            score += 1
    print("Error rate of QDA on training data set: " + str((1-(score/len(y_training)))*100) + "%") 
    print()
    
    #Accuracy
    score = 0
    for i in range(len(y_testing)):
        if (y_testing[i] != y_pred_test[i]):
            print("Error at index: " + str(i))
            print("Actual Data: " + y_testing[i])
            print("Predicted Data: " + y_pred_test[i])
            print()
        else:
            score += 1
    print("Error rate of QDA on testing data set: " + str((1-(score/len(y_testing)))*100) + "%")
    print()

#take in iris.data.txt file and turn into data
data = split_data(data_to_list('iris.data.txt'))

#assign training and testing data
t_data = data[0]
test_data = data[1]

#Run LDA/QDA analysis on all data
print("--------------------------------")
print("Testing on all attributes:")
LDA(t_data, test_data)
QDA(t_data, test_data)     

#list of attributes: sepal length, sepal width, petal length, petal width, class
#Running LDA and QDA analysis without each of the attributes 
print("--------------------------------")
print("Testing without sepal length:")
train1 = []
test1 = []
for x in t_data:
    y = x.copy()
    y.pop(0)
    train1.append(y)
for t in test_data:
    s = t.copy()
    s.pop(0)
    test1.append(s)

LDA(train1, test1)
QDA(train1, test1)
print("--------------------------------")
print("Testing without sepal width:")
train2 = []
test2 = []
for x in t_data:
    y = x.copy()
    y.pop(1)
    train2.append(y)
for t in test_data:
    s = t.copy()
    s.pop(1)
    test2.append(s)

LDA(train2, test2)
QDA(train2, test2)
print("--------------------------------")
print("Testing without petal length:")
train3 = []
test3 = []
for x in t_data:
    y = x.copy()
    y.pop(2)
    train3.append(y)
for t in test_data:
    s = t.copy()
    s.pop(2)
    test3.append(s)

LDA(train3, test3)
QDA(train3, test3)

print("--------------------------------")
print("Testing without petal width:")
train4 = []
test4 = []
for x in t_data:
    y = x.copy()
    y.pop(3)
    train4.append(y)
for t in test_data:
    s = t.copy()
    s.pop(3)
    test4.append(s)

LDA(train4, test4)
QDA(train4, test4)
print("--------------------------------")
```
