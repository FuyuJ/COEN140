```python
"""
Lab 7 - COEN 140
Marianne Fuyu Yamazaki Dorr
11/1/2022

Spam classification using logistic regression in Python. 
"""

import numpy as np
import pandas as pd
import math

def get_data(file):
    return pd.read_csv(file ,header = None)

train = get_data('spambase/spam-train')
test = get_data('spambase/spam-test')

#Training Data
#add training data except for output to x
x_train = train.drop(57, axis = 1)

#add output data to y, first column of train
y_train = train[57]

#Testing Data
#add testing data except for output to x
x_test = test.drop(57, axis = 1)

y_test = test[57]

def standardize_feature(data):
    standardized_data = (data-data.mean(axis=0))/data.std(axis=0)
    return standardized_data

def sigmoid(r):
    value = 1 /(1+math.exp(-r))
    return value

vectorized_sigmoid = np.vectorize(sigmoid)

class logistic_regression:
    def fit(self, xtrain, ytrain):
        x = xtrain.copy()
        x['57'] = np.ones(x.shape[0])
        
        w0 = np.random.normal(0, 1, size = x.shape[1])
        
        diff = 1
        while diff > 10e-5:
            
            w1 = w0 + 5*(10e-5) * np.dot(x.T, (ytrain - vectorized_sigmoid(np.dot(x,w0))))
            diff = np.linalg.norm(w1-w0)
            w0 = w1
        self.w0 = w0
        return self
    
    def pred(self, xtest):
        x = xtest.copy()
        x['57'] = np.ones(x.shape[0])
        predictions = vectorized_sigmoid(np.dot(self.w0, x.T))
        return predictions
    
log_reg = logistic_regression()
log_reg.fit(standardize_feature(x_train), y_train)

train_pred_prob = log_reg.pred(standardize_feature(x_train))
test_pred_prob = log_reg.pred(standardize_feature(x_test))

train_pred_value = []
test_pred_value = []

for p in train_pred_prob:
    if p > 0.5:
        train_pred_value.append(1)
    else:
        train_pred_value.append(0)

for x in test_pred_prob:
    if x > 0.5:
        test_pred_value.append(1)
    else:
        test_pred_value.append(0)

def error(y_actual, y_pred):
    count = 0
    for i in range(len(y_actual)):
        if y_pred[i] != y_actual[i]:
            count += 1
    error = count/len(y_actual)
    print(count)
    return error

training_error = error(y_train, train_pred_value)*100
print("Error rate of the training data:{}%".format(training_error))

testing_error = error(y_test, test_pred_value)*100
print("Error rate of the testing data:{}%".format(testing_error))

    
```
