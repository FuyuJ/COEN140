```python
"""
Lab 5 - COEN 140
Marianne Fuyu Yamazaki Dorr
10/18/2022

Implement  Linear  Regression  and  Ridge  Regression  in  Python.  
"""

import numpy as np
import pandas as pd
import math

def get_data(file):
    return pd.read_csv(file ,delimiter='\t')

def RMSE(actual, predicted):
    m = len(predicted)
    dif = (actual - predicted)
    sumsquared = 0
    
    for x in dif:
        sumsquared += x**2
    
    rmse = math.sqrt(sumsquared/m)
    return rmse

train = get_data("crime-train.txt")
test = get_data ("crime-test.txt")

#Training Data
#add training data except for output to x
x_train = train.drop('ViolentCrimesPerPop', axis = 1)
#add dummy column of 1s to solve for intercept b
dummy = np.ones(len(x_train))
x_train = np.column_stack((x_train, dummy))
#ensure that data is of correct type float64
x_train = np.float64(x_train)

#add output data to y, first column of train
y_train = train['ViolentCrimesPerPop']
#ensure that data is of correct type float64
y_train = np.float64(y_train)

#Testing Data
#add testing data except for output to x
x_test = test.drop('ViolentCrimesPerPop', axis = 1)
dummy2 = np.ones(len(x_test))
x_test = np.column_stack((x_test, dummy2))
x_test = np.float64(x_test)

y_test = test['ViolentCrimesPerPop']
y_test = np.float64(y_test)

"""
Linear Regression
"""
print("Problem 1 - Linear Regression")
#w = (xTx)-1xTy for inverse use np.linalg.inv(whatever you want to inverse)
# for dot equation use np.dot(matrixA, matrixB), for transpose use matrixA.T

def lin_reg(x, y):
    first = np.linalg.inv(np.dot(x.T, x))
    second = np.dot(x.T, y)
    w = np.dot(first, second)
    return w

w = lin_reg(x_train, y_train)

def problem1(x):
    #y = xTw
    return np.dot(x, w)

#find predicted y of training data
y = problem1(x_train)
print()
print("RMSE of Linear Regression Model on Training Data:")
print(RMSE(y_train, y))
print()
#find predicted y of testing data

y = problem1(x_test)

print("RMSE of Linear Regression Model on Testing Data:")
print(RMSE(y_test, y))
print()

"""
Ridge Regression
"""
print("Problem 2 - Ridge Regression")
def ridge_reg(x, y, l):
    #identity matrix must be of size (p+1) by (p+1)
    #(xTx + lambdaI)-1xTy
    #use np.identity(size) to create identity matrix
    lambda_matrix = l * np.identity(len(x_train.T))
    first = np.linalg.inv(np.dot(x.T, x) + lambda_matrix)
    second = np.dot(x.T, y)
    w = np.dot(first, second)
    return w

#Given Î» = 100
w2 = ridge_reg(x_train, y_train, 100)

def problem2(x):
    return np.dot(x, w2)

#find predicted y of training data
y = problem2(x_train)
print()
print("RMSE of Ridge Regression Model on Training Data:")
print(RMSE(y_train, y))
print()
#find predicted y of testing data

y = problem2(x_test)

print("RMSE of Ridge Regression Model on Testing Data:")
print(RMSE(y_test, y))
print()


```
